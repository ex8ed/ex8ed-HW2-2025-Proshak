### Just one second!

> Поскольку замер времени для всех методов имеет одну природу, я решил завернуть их в функции, к которым прикрутил декоратор для лаконичного и эффективного измерения времени: подпространство имен time, возможно, оверкилл, но по такой задумке код потенциально легко масштабировать, а еще проще понимать. Время, соответственно, возвращается в кортеже параметров последним значением во всех случаях.

```python
class what:
  class time:
    @staticmethod
    def is_(func):
      def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        tt = end - start
        print(f'{func.__name__} took {tt:.4f} seconds')
        return result, tt
      return wrapper
```

### Preprocessing moments

> Препроцессинг включает в себя регулярку для поиска слов на русском и английском, токенизатор и расширенный список стоп-слов из nltk. Почему-то в нем не хватает местоимений; pymorphy возможно подошел бы лучше для данной задачи, но окружение colab, кажется, его презирает из-за версии интерпретатора, поэтому пришлось прибегнуть к костылю с **custom_stop**.

```python
def preprocess_songs(songs, song_names):
    def clean_text(text):
        text = text.lower()
        text = re.sub(r'[^а-яa-z\s]', '', text)
        tokens = word_tokenize(text)
        custom_stop = [
            'это', 'этот', 'эта', 'эти', 'этой', 'этом', 'этому',
            'весь', 'вся', 'все', 'всё', 'всех', 'всем', 'всеми',
            'который', 'которая', 'которые', 'которого', 'которой',
            'еще', 'ещё', 'уже', 'будто', 'ведь', 'вон', 'вот',
            'даже', 'едва', 'именно', 'иной', 'как', 'какой',
            'когда', 'кто', 'либо', 'может', 'можно', 'надо',
            'наш', 'него', 'нее', 'них', 'нибудь', 'никогда',
            'однако', 'очень', 'почему', 'просто', 'себя',
            'снова', 'тоже', 'хорошо', 'чтобы', 'чуть', 'пока',
            'твой', 'твои', 'мой', 'мои', 'его', 'их', 'мной'
        ]
        stop_words = set(stopwords.words('russian') +
                         stopwords.words('english') +
                         custom_stop)
        tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
        return ' '.join(tokens)

    return [clean_text(song) for song in songs]
```

### All atention on the stage!

> Ниже просто имплементация функций, реализующих каждый из рассматриваемых методов

```python
@what.time.is_
def count_vec(txt):
  c_v = CountVectorizer()
  return c_v.fit_transform(txt), c_v.get_feature_names_out()


@what.time.is_
def tfidf(txt):
  tf = TfidfVectorizer(use_idf=True)
  return tf.fit_transform(txt), tf.get_feature_names_out()


@what.time.is_
def w2v(txt):
  sts = [tt.split() for tt in txt]
  return Word2Vec(sts,
                  vector_size=70,
                  window=5,
                  min_count=1,
                  workers=4)
```

```python
prepared_songs = preprocess_songs(songs, song_names)
```

```python
(count_vectorizer_matrix, feature_names), cv_time = count_vec(prepared_songs)
```

```python
(tf_idf_matrix, feature_names_tfidf), tf_time = tfidf(prepared_songs)
```

```python
word2vec_model, w2v_time = w2v(prepared_songs)
```

### Time's up!

> Визуализация времени работы каждого из методов позволяет сделать вывод о том, что все сработало предсказуемо: **CountVec** – чемпион, **tf-idf** уехал не сильно дальше, в рамках статистической погрешности, и, как и ожидалось, **word2vec** занял наибольшее время выполнения в силу архитектуры и нейро-начинки.

```python
times = [cv_time, tf_time, w2v_time]
methods = ['CountVectorizer', 'TF-IDF', 'Word2Vec']
colors = ['skyblue', 'lightcoral', 'lightgreen']

plt.figure(figsize=(8, 5))
bars = plt.bar(methods, times, color=colors)

for bar, t in zip(bars, times):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
             f'{t:.4f} с', ha='center', va='bottom', fontsize=10)

plt.ylabel('Время (секунды)')
plt.title('Сравнение времени работы методов векторизации')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

### Nice to meet ya, pal

> Статистика самых частых слов для **CountVec** и **TF-IDF** отличается: **TF-IDF**, исходя из структуры текста, действительно уделяет внимание на распределенность слов по всему тектовому корпусу: **CountVec** заметно делает акцент на одном из текстов ("ножевое"), выводя в топ-3 слова наиболее встречающиеся в нем элементы, в то время как **TF-IDF** выделяет наиболее распределенные примеры. В остальном топ наиболее встречающихся слов похож.
```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

word_freq_count = {feature_names[i]: count_sum[i] for i in range(len(feature_names)) if count_sum[i] > 0}
wc_count = WordCloud(width=800, height=400,
                     background_color='white',
                     colormap='Blues',
                     max_words=50).generate_from_frequencies(word_freq_count)
axes[0].imshow(wc_count, interpolation='bilinear')
axes[0].axis('off')
axes[0].set_title('CountVectorizer', fontsize=14, fontweight='bold')


word_freq_tfidf = {feature_names_tfidf[i]: tfidf_sum[i] for i in range(len(feature_names_tfidf)) if tfidf_sum[i] > 0}
wc_tfidf = WordCloud(width=800, height=400,
                     background_color='white',
                     colormap='Reds',
                     max_words=50).generate_from_frequencies(word_freq_tfidf)
axes[1].imshow(wc_tfidf, interpolation='bilinear')
axes[1].axis('off')
axes[1].set_title('TF-IDF', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Hot, hot, hot!

> Визуализированная частота по всем словам в текстовом корпусе в форме хитмапы позволяет выделить признаки встречаемости слов для каждого из методов. Несмотря на то, что **CountVec** склонен выделять несколько цветовые пики, выбирая просто наиболее встречающуюся форму, на данном текстовом корпусе подобных результатов немного: художественная и музыкальная формы, выбранные авторами, не включают в себя частое повторение, из чего следует практически равномерный окрас по всему множеству слов как при применении **TF-IDF**, так и **CountVec**. Однако пики распределены у методов по разному, что позволяет подтвердить утверждение о том, что **CountVec** собирает общую статистику, в то время как **TF-IDF** старается опираться на контекст.

> Средние эмбеддинги также сглажены и не демонстрируют резких всплесков. Кластеризировать полученные слова по смысловому значению на малом количестве кластеров скорее всего не выйдет, однако, следует предположить, что при применении метрики **t-sne** некоторые слова могут стоять рядом по смысловому признаку. Текстовых данных слишком мало и они выполнены с акцентом на разнообразие формы, поэтому сделать дополнительные однозначные выводы проблематично.

### So, what are we talking about?

> **t-sne** при большем количестве слов в корпусе позволило бы сгруппировать их по смысловому контексту, опираясь на эмбеддинги, однако на графике, приведенном ниже, можно убедиться в том, что выделять кластеры проблематично. За кулисами я поэкспериментировал с **KMeans**, что не привело к значимым результатам: при малом количестве кластеров слова группировались совершенно нерелевантным образом (что-то вроде "книги", "зубами", "городу"), и, как следствие, сделать значимые аналитические выводы нельзя.

> Более того, не только кластеры, но и сама близость на плоскости не всегда гарантирует семантическую схожесть выделенных слов.

```python
N = 50

words = list(word2vec_model.wv.index_to_key)
word_freq = [(word, word2vec_model.wv.get_vecattr(word, 'count')) for word in words]
word_freq.sort(key=lambda x: x[1], reverse=True)
top_words = [w for w, _ in word_freq[:N]]

top_vectors = np.array([word2vec_model.wv[word] for word in top_words])

perplexity = min(30, N-1)
tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity)
vectors_2d = tsne.fit_transform(top_vectors)

plt.figure(figsize=(14, 10))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.6, s=50)

for i, word in enumerate(top_words):
    plt.annotate(word, (vectors_2d[i, 0], vectors_2d[i, 1]),
                fontsize=9, alpha=0.8)

plt.title(f't-SNE визуализация Word2Vec (топ-{N} слов по частоте)')
plt.xlabel('t-SNE компонента 1')
plt.ylabel('t-SNE компонента 2')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
